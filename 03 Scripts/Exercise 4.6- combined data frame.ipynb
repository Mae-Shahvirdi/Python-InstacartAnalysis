# 03. Determining a suitable to combine the 2 dataframes

#### A suitable way to combine the df_ords_prods_combined data frame with the df_prods data frame, despite their different shape, is to merge them thanks to their shared column "product_id". For this project, a default inner join is advisable.

```python
# Merge df_ords_prods_combined with df_prods (inner join)
df_merged =  df_prods.merge(df_ords_prods_combined, on= 'product_id', indicator = True)
```
