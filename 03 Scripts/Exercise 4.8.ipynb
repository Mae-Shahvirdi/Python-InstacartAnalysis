# 03. Create a loyalty flag for existing customers using transform() and loc() functions

```python
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform('max')
```

```python
ords_prods_merge.head()
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] >40, 'loyalty_flag'] = 'loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <=40) & (ords_prods_merge['max_order'] >10), 'loyalty_flag'] = 'regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order']<= 10, 'loyalty_flag'] = 'new customer'
```

```python
# check the frequency
ords_prods_merge['loyalty_flag'].value_counts(dropna= False)
```

# 05. Create a spending flag for each user based on the average price across their orders using this criteria:
### If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
### If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

```python
ords_prods_merge['mean_product_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform('mean')
```

```python
ords_prods_merge[['user_id', 'mean_product_price']].head()
```

```python
ords_prods_merge.loc[ords_prods_merge['mean_product_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['mean_product_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
ords_prods_merge['spending_flag'].value_counts(dropna= False)
```

```python
ords_prods_merge[['user_id', 'mean_product_price', 'spending_flag']].head()
```

##### The number of Low spender is significiantly lesser than the number of high spenders

# 0.6- Determine the frequency of customers by using an order frequency flag to mark the regularity of the user's ordering behaviors according to the median in the 'days_since_prior_order' column.
### If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
### If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
### If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
ords_prods_merge['median_days_since_prior_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform('median')
```

```python
ords_prods_merge[['user_id', 'days_since_prior_order', 'median_days_since_prior_order']].head()
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order'] > 20, 'order_frequency'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_days_since_prior_order'] <= 20) & (ords_prods_merge['median_days_since_prior_order'] > 10), 'order_frequency'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order'] <= 10, 'order_frequency'] = 'Frequent customer'
```

```python
ords_prods_merge['order_frequency'].value_counts(dropna = False)
```

##### There are 5 nan values, which should be for the users who has not yet ordered but they were counted as customers.
